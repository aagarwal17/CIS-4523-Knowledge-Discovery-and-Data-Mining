### Task 3: Scrape the NASA Data:

```python
#We now scrape the NASA data (http://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html) to get additional 
#features/information about these solar flares

#Using requests to get page content, similar to Task 1:
r2 = rq.get('http://cdaw.gsfc.nasa.gov/CME_list/radio/waves_type2.html')
#Extracting the text from the page:
theText = r2.text
#Using BeautifulSoup to read and parse the data as html:
mySoup = bs(theText, 'html.parser')
#Using prettify to view the content and find the appropiate table
mySoup.prettify()
```

As can be seen, we need to remove the unnecessary first 12 rows as well as the last 2 rows.

```python
#Removing the first 12 unnecessary lines
for i in range(0,12):
    dataRows.pop(0);
#Remove the last two useless lines
dataRows.pop(len(dataRows)-1)
dataRows.pop(len(dataRows)-1)

dataRows
```

```python
# Creating the Dataframe with the data in dataRows array. Names are chosen for the columns as suggested in the
#example in the instructions:

# Making an empty table:
tableNASAdf = pd.DataFrame(columns = ['start_date', 'start_time', 'end_date', 'end_time', 'start_frequency', 'end_frequency', 'flare_location', 'flare_region','flare_classification', 'cme_date', 'cme_time', 'cme_angle', 'cme_width', 'cme_speed'], index = range(0,len(dataRows)))

#Filling the table by iterating through each line in dataRows:
#Creating a row counter, which is helpful for accurately inputting the data into the table, 'tableNASAdf'
rowNum = 0
for line in dataRows:
    #Split the data wherever there is a space for that 'line':
    features = line.split(' ')
    #After running the code once, I realized that there was extra spacing adding in certain areas to line 
    #the data up evenly.Thus, we remove any spaces being treated as extra 'features' that were created by 
    #the line.split(' ') above.
    while '' in features:
        features.remove('')
    #Adding the data to the DataFrame, tableNASAdf:
    columnNum = 0
    while (columnNum < 14):
        #After examining the dataset, it appears that column 15, the plots, were added after this assignment 
        #was written.Therefore, to adhere to the guidelines of the assignment, we ignore this last column 
        #(and any other extra information that appears after it), which does not seem to be important for our 
        #purposes anyways. We see later in task 4 that the example dataset preview includes the plot column, 
        #but I still believe this column will not help us with the rest of our assignment; therefore, 
        #I will keep this column out still.
        if (columnNum == 15):
            break
        tableNASAdf.iat[rowNum, columnNum] = features[columnNum]
        columnNum += 1
    rowNum += 1
tableNASAdf
```

### Task 4: Tidy the NASA Table:

```python
#First, we recode any missing entries at NaN. After looking carefully at the NASA data and the data descriptions, 
#I have found that:
#---- and ----- are missing data entries that can exist in multiple columns (for Flare and CME data), 
#???? is a missing data entry that can appear in the frequency columns and indicates that the starting 
#and ending frequencies are not determined,
# --/-- is a missing data entry that can appear in the date column for CME,
# and --:-- is a missing data entry that can appear in the time column for CME.
#Replacing these missing entries:
tableNASAdf = tableNASAdf.replace('----', np.nan)
tableNASAdf = tableNASAdf.replace('---', np.nan)
tableNASAdf = tableNASAdf.replace('-----', np.nan)
tableNASAdf = tableNASAdf.replace('????', np.nan)
tableNASAdf = tableNASAdf.replace('--/--', np.nan)
tableNASAdf = tableNASAdf.replace('--:--', np.nan)
tableNASAdf = tableNASAdf.replace(to_replace = '360h', value = '360') #There was a typo in the table in which an h was added

tableNASAdf
```

```python
#The CPA column (cme_angle) contains angles in degrees for most rows, except for halo flares, 
#which are coded as Halo. 
#I create a new column that indicates if a row corresponds to a halo flare or not, 
#and then replace Halo entries in the cme_angle column with NaN:

#Creating and filling new 'is_halo' column as indicated:
tableNASAdf['is_halo'] = tableNASAdf['cme_angle'].map(lambda x: x == 'Halo')

#Replacing "Halo" with 'NaN'
tableNASAdf = tableNASAdf.replace('Halo', np.nan)

tableNASAdf
```

```python
#A helper function to use in map function below to determine if a given string is an integer/number:
def isInteger(string):
    try: 
        #Check if the string is a float because NaN values are floats; integers will work here as well
        float(string)
        return True
    except ValueError:
        return False
```

```python
#The width column indicates if the given value is a lower bound. 
# I create a new column that indicates if width is given as a lower bound, and remove any non-numeric part 
#of the width column:

#Creating and filling new 'width_lower_bound' column as indicated:
tableNASAdf['width_lower_bound'] = tableNASAdf['cme_width'].map(lambda x: str(x)[0] == '>')

#Removing any non-numeric part of the width column:
tableNASAdf['cme_width'] = tableNASAdf['cme_width'].map(lambda x: x if isInteger(x) else x[1:])

#Here, I display the first 50 columns to clearly demonstrate that we have completed the task as requested.
tableNASAdf.head(50)
```

```python
#Finally, we combine the data and time columns for start, end, and cme so they can be encoded as datetime objects:

#The to_datetime method runs into issues with the times that say 24:00 in the NASA dataset 
#(as the interval should only be between 0:00 and 23:59, and 24:00 should be 0:00 of the next day),
#so rather than deal with all the various cases before converting the time to a date time format 
#(for example, the day being incremented by one may lead to a new month with some months and not others 
#because some months have 30 days and others have 31), I change the 24:00 times in the dataset to 23:59, 
#and will increment the time by 1 second once it is in a to_datetime format.
#It can be noted that, after looking closely at the dataset, all 24:00 times exist in the 'end_time' column, 
#but it will make no diffrence to do the replace across the entire dataset, as done below:
tableNASAdf = tableNASAdf.replace(to_replace = '24:00', value = '23:59')
tableNASAdf.head(70)

# Combining the date column with the time columns to convert to a datetime format
tableNASAdf_st = pd.to_datetime(tableNASAdf['start_date'] + ' ' + tableNASAdf['start_time'])
tableNASAdf_et = pd.to_datetime(tableNASAdf['start_date'].str[:5] + tableNASAdf['end_date'] + ' ' + tableNASAdf['end_time'])
tableNASAdf_cmet = pd.to_datetime(tableNASAdf['start_date'].str[:5] + tableNASAdf['cme_date'] + ' ' + tableNASAdf['cme_time'])
```

```python
#Converting 23:59 times back to 24:00 times for each of our 3 lists.
#We note that the dataset did not consists of any 23:59 times before we set the 24:00 values to 23:59, 
#so we are not falsifying any data
for dateTime in range(len(tableNASAdf_st)):
    if (str(dt.datetime.strptime(str(tableNASAdf_st[dateTime]), "%Y-%m-%d %H:%M:%S").time()) == "23:59:00"):
        #print(tableNASAdf_st[dateTime])
        tableNASAdf_st[dateTime] = tableNASAdf_st[dateTime] + dt.timedelta(0,60)
        #print(tableNASAdf_st[dateTime])
for dateTime in range(len(tableNASAdf_et)):
    if (str(dt.datetime.strptime(str(tableNASAdf_et[dateTime]), "%Y-%m-%d %H:%M:%S").time()) == "23:59:00"):
        #print(tableNASAdf_et[dateTime])
        tableNASAdf_et[dateTime] = tableNASAdf_et[dateTime] + dt.timedelta(0,60)
        #print(tableNASAdf_et[dateTime])
for dateTime in range(len(tableNASAdf_cmet)):
    if ((str(tableNASAdf_cmet[dateTime]) != 'NaT') and str(dt.datetime.strptime(str(tableNASAdf_cmet[dateTime]), "%Y-%m-%d %H:%M:%S").time()) == "23:59:00"):
        #print(tableNASAdf_cmet[dateTime])
        tableNASAdf_cmet[dateTime] = tableNASAdf_cmet[dateTime] + dt.timedelta(0,60)
        #print(tableNASAdf_cmet[dateTime])
```

```python
# Adding the new columns of the datetimes to tableNASAdf
tableNASAdf['start_datetime'] = tableNASAdf_st
tableNASAdf['end_datetime'] = tableNASAdf_et
tableNASAdf['cme_datetime'] = tableNASAdf_cmet

# Droping the old columns
tableNASAdf = tableNASAdf.drop('start_date', 1)
tableNASAdf = tableNASAdf.drop('start_time', 1)
tableNASAdf = tableNASAdf.drop('end_date', 1)
tableNASAdf = tableNASAdf.drop('end_time', 1)
tableNASAdf = tableNASAdf.drop('cme_date', 1)
tableNASAdf = tableNASAdf.drop('cme_time', 1)

tableNASAdf
```

```python
#Rearranging the columns to match the order seen in the example 
#(NOTE: we use different names for some of them for consistency with our past code, 
#and we do not include the plot attribute for the reason stated earlier):
tableNASAdf = tableNASAdf[['start_datetime', 'end_datetime', 'start_frequency', 
                       'end_frequency', 'flare_location', 'flare_region','flare_classification', 
                       'cme_datetime', 'cme_angle', 'cme_width', 'cme_speed', 'is_halo', 
                       'width_lower_bound']]
tableNASAdf
```

### Task 6: Integration:

```python
#For Task 6, we write a function that finds the best matching row in the NASA data for each of the top 50 
#solar flares in the SpaceWeatherLive data.
# As stated in the instructions, I must decide for myself what is the best matching entry in the NASA data 
#for each of the top 50 solar flares.
#I am defining the best matching rows across the two datasets based on the flare_classification, start_datetime, 
#and end_datetime.
#We first note that we can only use attributes that exist in both datasets, which are flare classfication, 
#start datetime, end datetime, and region (we are not confident cme_datetime = max_datetime).
#Since some of the region codes for the NASA dataset seem to be similar to ones in the SWL dataset but just 
#add a '10' at the beginning, I believe NASA uses a different system to write their region codes.
#Since we do not know how the NASA region code system works, we cannot transform them to be usable with the 
#SWL dataset. Thus, region codes are not good determiners for the best match.
#Now, we use flare_classification because it can be transformed into an integer value (which is easier to work with), 
#and it is pretty unique (it is rare that two rows in the SWL data have the same classfication).
#We will store multiple close flare_classification values for a given row in the SWL dataset, and then choose 
#the best matching one by looking at that start_datetime and end_datetime.
#Thus, we use start_datetime and end_datetime to shorten our choices to just 1, which will be the best match 
#in the NASA data for the respective row in the SWL data.

#Making a copy of NASA data table. We add a column to the table, called SWLRank to indicate the rank according 
#to SpaceWeatherLive, if it appeared in that dataset.
#We set these values to NaN to start, but we will update them to the appropiate rank in the functions below.
out_NASA = tableNASAdf.copy()
out_NASA['SWLRank'] = pd.Series(np.nan, index=out_NASA.index)

#Making a copy of the SWL table. We edit the 'x_classification' column to have floats by stripping the
#'X' and '+' and casting.
#Since this table is just a copy, the original is left unchanged, and there is no problem in doing this.
out_SWL = myTable.copy()
out_SWL['x_classification'] = out_SWL['x_classification'].str.lstrip('X')
out_SWL['x_classification'] = out_SWL['x_classification'].str.rstrip('+').astype(float)

#In task 5, we obtained the rows that had a flare classification with X and stored it in a new dataframe 
#using the below line:
topFlaresdf2 = tableNASAdfCopy.loc[tableNASAdfCopy['flare_classification'].str.contains('X')]
#Since SWL only has flare classification's with X, when doing our comparisons for the best match, we can 
#just look at the rows from the NASA dataset with the 'X', which are all in totalFlaresdf2

#Similar to above, we get rid of the 'X' character and cast the column to float type so that we can compare 
#this column to the respective column for the SWL data.
topFlaresdf2['flare_classification'] = topFlaresdf2['flare_classification'].str.lstrip('X').astype(float)


#A helper list that will store the indicies of the rows from the NASA dataset that we no longer want to use in 
#comparison for the rest of the SWL data rows because they were already chosen as best matches for other SWL data rows.
listToDrop = []

#A helper function to find the nearest time
def nearestTime(items, pivot):
    return min(items, key=lambda x: abs(x - pivot))

#My function to find the best match. It takes a row from the SWL dataset as an argument.
def bestMatch(row):    
    tolerance = 10 #Tolerance is continuously updated value that is used to shorten down our choices for the 
    #closest classification scores (and their respective rows)
    
    #If listToDrop has indicies in it, there are rows that need to be dropped from our evaluation dataset, 
    #which we do in the if block.
    #Otherwise, we are still at the first iteration, so we do not have any values in listToDrop. 
    #Thus, we just take a copy of topFlaresdf2
    if (len(listToDrop) > 0):
        topFlaresdf2Temp = topFlaresdf2.copy()
        topFlaresdf2Temp = topFlaresdf2Temp.drop(listToDrop, axis = 0)
    else:
        topFlaresdf2Temp = topFlaresdf2.copy()
    #While our temp NASA dataset still has more than 5 rows in it, we try to remove rows that are not close 
    #to the inputted SWL row, based on the flare classification.
    #We chose 5 arbitrarily, but it seems to be a good amount to keep before we shorten down the list further 
    #when comparing the start_datetime and end_datetime.
    while (len(topFlaresdf2Temp.index) > 5):
        #for each row in the temp NASA dataset, if the absolute difference of the flare classifications from 
        #the SWL and NASA data rows is greater than the given tolerance,
        #then that NASA dataset row is too far from the SWL data row to be the best match, so remove it from 
        #the temp NASA dataset. 
        for j, val in topFlaresdf2Temp.iterrows():
            if (abs(val['flare_classification'] - row['x_classification']) >= tolerance):
                topFlaresdf2Temp = topFlaresdf2Temp.drop(j, axis = 0)
        tolerance -= .1 #The tolerance value starts off large to make sure we get at least 5 rows for the best match. 
        #We decrement the tolerance by .1 to slowly get rid of more NASA data rows that are not close enough.    
    
    #Printing the current flare class value from SWL data for demonstration/understanding
    print("Current Flare Class Value from SWL: ", row['x_classification']) 
    #Printing the closest row matches found, based on the flare value
    print("Closest Matches Found: \n", topFlaresdf2Temp['flare_classification']) 
    
    #Now, we shorten the list of 5 potential best match rows from the NASA dataset using the 
    #start_datetime and end_datetime.
    SWLrowStartTime = row['start_datetime'] #Storing the current SWL data row start_time in a variable
    #Finding the nearest time to the SWL start time using the helper function 'nearestTime', and storing it in a variable.
    closestStartTime = nearestTime(topFlaresdf2Temp['start_datetime'], SWLrowStartTime)
    #Printing the current start datetime value from SWL data for demonstration/understanding
    print("Current Start DateTime from SWL: ", SWLrowStartTime)
    #Printing the closest start datetime to the SWL datetime
    print("Closest Start DateTime from NASA: ", closestStartTime) 
    
    SWLrowEndTime = row['end_datetime'] #Storing the current SWL data row end_time in a variable
    #Finding the nearest time to the SWL end time using the helper function 'nearestTime', and storing it in a variable.
    closestEndTime = nearestTime(topFlaresdf2Temp['end_datetime'], SWLrowEndTime)
    #Printing the current end datetime value from SWL data for demonstration/understanding    
    print("Current Start DateTime from SWL: ", SWLrowEndTime)
    #Printing the closest end datetime to the SWL datetime    
    print("Closest End DateTime from NASA:  ", closestEndTime)
    
    #With the closest start and end time options, we iterate through our short list of potential best match rows 
    #from the NASA dataset, and drop the ones that do not match these times.
    for k, value in topFlaresdf2Temp.iterrows():
        if (value['start_datetime'] != closestStartTime or value['end_datetime'] != closestEndTime):
            topFlaresdf2Temp = topFlaresdf2Temp.drop(k, axis = 0)
    print("The Best Match: \n", topFlaresdf2Temp) #Printing the best match, based on my method of finding the best match
        
    index = np.nan #Creating an index variable to use for the below for-loop
    #I needed the index for the best match, and I could not seem to access it without looping 
    #through the 1-row NASA dataset
    for l, val in topFlaresdf2Temp.iterrows():
        index = l
    #The index for the best match is used to find the matching row in the full NASA (cleaned) dataset. 
    #This row's SWLRank value is then updated to be the rank value for the SWL data row we are currently working with    
    out_NASA.at[index, 'SWLRank'] = row['rank']
    #We add the index to list to drop because this index represents the row from the NASA dataset we 
    #identified as a best match for a row in the SWL dataset, so it needs to be removed from further evaluation.    
    listToDrop.append(index) 

#The for loop to iterrate through the 50 rows of the SWL dataset and find the respective best match in the NASA dataset
for i, row in out_SWL.iterrows():
    bestMatch(row)
```

### Task 9: Events Distribution:

```python
#Making a new dataframe with the count of strong flares (top 50) per month 
#(based on start_datetime; I could have chosen end_datetime but either is fine).
top50Month = topFlaresdf['start_datetime'].groupby(topFlaresdf['start_datetime'].dt.to_period("M")).agg('count')
top50Monthdf = top50Month.to_frame()
top50Monthdf = top50Monthdf.rename(columns = {'start_datetime': 'count'})
top50Monthdf = top50Monthdf.reset_index()
top50Monthdf['start_datetime'] = top50Monthdf['start_datetime'].dt.to_timestamp()
top50Monthdf
```

```python
#Making a new dataframe with the count of all flares (all NASA data) per month 
#(based on start_datetime; I could have chosen end_datetime but either is fine).
AllMonth = out_NASA['start_datetime'].groupby(out_NASA['start_datetime'].dt.to_period("M")).agg('count')
AllMonthdf = AllMonth.to_frame()
AllMonthdf = AllMonthdf.rename(columns = {'start_datetime': 'count'})
AllMonthdf = AllMonthdf.reset_index()
AllMonthdf['start_datetime'] = AllMonthdf['start_datetime'].dt.to_timestamp()
AllMonthdf
```

```python
#Creating the plot
mp.pyplot.figure(figsize = (10,6), dpi = 80)
mp.pyplot.title('Number of Flares Per Month Over Time for NASA Dataset')
mp.pyplot.xlabel('Month-Year')
mp.pyplot.ylabel('Number of Flares')
mp.pyplot.plot(AllMonthdf['start_datetime'], AllMonthdf['count'], '*', label = 'All NASA data')
mp.pyplot.plot(top50Monthdf['start_datetime'], top50Monthdf['count'], 'go', label = 'Top 50')
mp.pyplot.legend(loc = 'upper right')
```

The above plot shows the number of flares per month over time, with the green dots indicating the counts for the strong flares (top 50). In general, it appears that the strong flare counts are smaller than counts seen when looking at the entire dataset. This makes sense considering there is a larger number of points being counted for the entire dataset, so the counts per month should be higher. The strong flares are also more clustered towards the time periods of about 1997-2007 as well as 2011-2017. The larger of these two clusters is between 1997-2007. Unfortunately, this pattern does not seem to be unique to the strong flares as those time-period clusters are basically seen for the entire dataset as well (clusters of about 1997-2008 and 2011-2020).
