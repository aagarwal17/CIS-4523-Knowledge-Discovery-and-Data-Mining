A set of nucleotides arranged in a certain order is biological sequence data (which falls under multivariate discrete sequence data in the dependency-oriented data category). Here, the sets may relate to strings of nucleotides or amino acids. The 'certain order' of these units provides information about the characteristics of protein function. The data mining process can then be used to determine insightful patterns that are reflective of varying biological properties. This type of discrete sequence data is more challenging for mining algorithms to deal with because of the lack of continuity that time-series data contains, for example. 

(b)  It is desired to partition customers into similar groups on the basis of their demographic profile. Which data mining problem is best suited to this task?

The data mining problem that is best suited to this task is clustering. Clustering is the task of discovering groups and structures in the data that are in some way or another "similar", without using known structures in the data. Here, we do not know the relations between the customers, but we are trying to partition them into similar groups based on their demographic profile. Thus, clustering will help to determine the relational similarities between the various demographical characteristics of the customers and group them as best as it can. This is unsupervised learning.

(c)  Suppose in problem 1.b, the merchant already knows for some  of the customers whether or not they have bought widgets. Which data mining problem would be suited to the task of identifying groups among the remaining customers, who might buy widgets in the future?

Classification would be the data mining problem best suited to the task of identifying groups among the remaining customers, who might buy widgets in the future. Classification is that task of generalizing known structure to apply to new data. Here, we know, for a portion of the dataset, that those customers have a particular characteristic (buying a widget). The characeristic is observed. Classification will help to determine who might buy widgets in the future by extrapolating the relationship seen in the portion of the dataset to the larger dataset. This is supervised learning.

```python
from scipy.stats import gaussian_kde as kde
from matplotlib.colors import Normalize
from matplotlib import cm

def makeColors(vals):
    colors = np.zeros((len(vals),3))
    norm = Normalize(vmin=vals.min(), vmax=vals.max())
    colors = [cm.ScalarMappable( norm=norm, cmap='jet').to_rgba(val) for val in vals]
    return colours

age = heartDataComplete['age'].tolist()
cp = heartDataComplete['cp'].tolist()
vals = [age, cp]

densObj = kde(vals)
colors = makeColors(densObj.evaluate(vals))

plt.figure(figsize = (10,6), dpi = 80)
plt.scatter(heartDataComplete['age'], heartDataComplete['cp'], color=colors)
plt.title('Scatter Plot of Chest Pain Type vs Age')
plt.xlabel('Age')
plt.ylabel('Chest Pain Type')
plt.yticks([1.0,2.0,3.0,4.0])
plt.show()
```

### Problem 3:

You are given a set of m objects that is divided into K groups, where the i-th group is of size mi . If the goal is to obtain a sample of size n < m , what is the difference between the following two sampling schemes? (Assume sampling with replacement.)

a.     We randomly select n ∗ mi /m elements from each group.

Randomly selecting n ∗ mi /m elements from each group would be stratified sampling, which means that the sample from each group is proportional to its size relative to the total number of objects. We guarentee that we have at least one observation from each of our K groups.

b.     We randomly select n elements from the data set, without regard for the group to which an object belongs.

Randomly selecting n elements from the data set--without regard for the group to which an object belongs--is just random sampling, which means each sample has an equal probability of being chosen. Here, the only way to guarentee that we have an observation from each group is if we have m elements. However, that is not allowed because we want n < m elements.

Difference: The first sampling scheme is guaranteed to get the same number of objects from each group, while the second sampling scheme does not get this guarentee (the number of objects from each group can and will vary). This is because the second sampling scheme only guarantes that, on average, the number of objects from each group will be n ∗ mi/m. Thus, the first sampling scheme will have less variance than the second randoming scheme since the latter also includes the variance between the different groups. Also, if the objects in each group are alike, then the first sampling technique guarentees a sample from each group, which can be used in analysis whenever the sample size is large enough. These advantages of the stratified sampling make it generally more accurate than the second sampling scheme's simple random sampling.

Find a PCA package and use it to compute eigenvectors and eigenvalues for this image.

a.     Compute 2, 5, and 10 principal components and show original and the resulting images.

b.     What is the minimal number of principal components needed to retain 80% of data variance?

```python
#Imports:
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
#Original Image:
img = plt.imread('hw2_2022_problem4_Face.png')
img = img/255
plt.imshow(img, cmap = 'gray')
```

```python
# First, I compute 2 principal components and show the resulting image:
pca_2 = PCA(n_components = 2)
pca_2.fit(img)
components_2 = pca_2.transform(img)
projected_2 = pca_2.inverse_transform(components_2)

fig = plt.figure()
ax1 = fig.add_subplot(1,2,2)
ax1.imshow(projected_2.reshape(112, 92), cmap = 'gray')
ax2 = fig.add_subplot(1,2,1)
ax2.imshow(img, cmap = 'gray')
#Note, I find the eigenvalues and eigenvectors later below
```

As expected the image is very blurry compared to the original with only two principal components.

```python
# Next, I compute 5 principal components and show the resulting image:
pca_5 = PCA(n_components = 5)
pca_5.fit(img)
components_5 = pca_5.transform(img)
projected_5 = pca_5.inverse_transform(components_5)


fig = plt.figure()
ax1 = fig.add_subplot(1,2,2)
ax1.imshow(projected_5.reshape(112, 92), cmap = 'gray')
ax2 = fig.add_subplot(1,2,1)
ax2.imshow(img, cmap = 'gray')

#Note, I find the eigenvalues and eigenvectors later below
```

It is clear that the image is much more visible with 5 principal components in comparison to the original image

```python
# Finally, I compute 10 principal components and show the resulting image:
pca_10 = PCA(n_components = 10)
pca_10.fit(img)
components_10 = pca_10.transform(img)
projected_10 = pca_10.inverse_transform(components_10)

fig = plt.figure()
ax1 = fig.add_subplot(1,2,2)
ax1.imshow(projected_10.reshape(112, 92), cmap = 'gray')
ax2 = fig.add_subplot(1,2,1)
ax2.imshow(img, cmap = 'gray')

#Note, I find the eigenvalues and eigenvectors later below
```

As expected, the image is a lot more clear, and much more comparable to the original image

```python
#Performing PCA on 2 components:
pca_Image = PCA(n_components = 2)
principalComponentsImage2 = pca_Image.fit_transform(x)
```

```python
#Outputting the dataset with 2 principal components:
principalComponentsImagedf2 = pd.DataFrame(data  = principalComponentsImage2, columns = ['principal  component 1', 'principal component 2'])
principalComponentsImagedf2
```

```python
print('Explained Variation per principal component: {}'.format(pca_Image.explained_variance_ratio_))
```

```python
#Centering the data and computing the sample covariance matrix:
x_centered = x - np.mean(x, axis = 0)
cov_matrix = np.dot(x_centered.T, x_centered)/112
eigenvalues = pca_Image.explained_variance_
for eigenvalue, eigenvector in zip(eigenvalues, pca_Image.components_):
    print("Eigenvalue through Method 1:", np.dot(eigenvector.T, np.dot(cov_matrix, eigenvector)))
    print("Eigenvalue through Method 2:",eigenvalue)
print("Eigenvectors:\n", pca_Image.components_)
```

Here, I print out the two eigen values twice because I found two different methods to find the eigen values, and they produce slightly different results. I kept both in my code for reference.

The TA told me during her office hours that she would like me to use these eigen vectors to reconstruct an array similar to the original (before PCA). To find this PCA reconstruction, we multiply the PCA scores by the eigen vectors and add the mean. First, we multiply the original normalized array/matrix of size 112 x 92 by the eigen vector matrix of size 92 x 2. This will produce a 112 x 2 matrix:

```python
print(type(pca_Image.components_))
print(type(x))
Z = np.dot(x, pca_Image.components_.T)
Z
```

```python
#We now multiply our 112 x 2 matrix by pca_Image.components (2 by 92) to get a 112 by 92 matrix, as we originally had:
X_hat = np.dot(Z, pca_Image.components_)

#Finally, we add the mean vector to get the final reconstruction:
mu = np.mean(x, axis = 0)
X_2Comp = X_hat + mu
X_2Comp
```

```python
#I now put the reconstructed array into a dataset to be more easily visible for comparison to the original:
feat_cols = ['feature' + str(i) for i in range(X_2Comp.shape[1])]
reconstructed2df = pd.DataFrame(X_2Comp, columns = feat_cols)
reconstructed2df
```

b.     Using a naive solution could solve the missing data problem. What are the advantages/disadvantages of this solution?

```python
#The naive solution is to drop the rows with missing values:
gym_naive = gym.dropna()
gym_naive
```

As stated above, a naive method of handling the null values in the dataset is to simply remove any rows with NaN values. The benfits are that the program would run more quickly since there would be less observations (as shown in the dataset above, we now have 671 rows instead of 5000), and as long as we have enough observations to not significantly reduce the ability of the program to complete its purpose (which it does not in this case since we still have 671 observations), our results should be consistent to having a larger dataset of full values. Nevertheless, this is also a huge disadvantage because losing observations means losing variation in the data that could be key to training a good model/gaining valuable information. We could have also removed the three columns with the missing data, but then we could not do as much computation (or any for our purposes). 

c.     Propose a better solution to solve the missing data problem.

A better solution/approach would be to fill these missing values in some form, of course. Specifically, we can start by finding the values for membership_type because this is easier than the other two columns. Which type of member a person is seems to be correlated to his or her age--based on my examination of the dataset--which makes sense considering the labels are "kids", "youths", and "adults". After discovering the mininum ages that show up in the initial dataset of youths and adults, I was able to relabel those values in the dataset. 

Next, we fill the missing values for the height and weight columns. When considering the various key techniques for filling nan values, I originally thought that it would be best (or at least easiest) to fill the values with the mean/average for that column, and continue forward with the analysis. However, I later noticed that age, sex, height, and weight have a correlation, which is why we can have a bmi as a valid indicator for various things. Thus, I created a heatmap showcasing the correlations, and we can see that age, sex, height, and weight are correlated, as hypothesized. Thus, a better solution for these columns would be to impute these values based on a regression framework.

```python
#Heat Map showcasing an existing correlation between age, sex, height, and weight:
var_corr = gym_naive.corr()
plt.figure(4, figsize = (8, 6))
sns.heatmap(var_corr, xticklabels = var_corr.columns, yticklabels = var_corr.columns, annot = False)
```

Since we see higher correlation colors/values between age, sex, height, and weight, we believe them to be correlated and will use this knowledge to make a regression framework to better fill the missing values. 

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# Evaluating Model Fit for imputation: heights
lm = linear_model.LinearRegression()
missing_naive = gym_naive[['age', 'sex', 'height', 'weight']]
x_height = missing_naive.drop('height', axis = 1)
y_height = pd.DataFrame(missing_naive['height'])
x_height_train, x_height_test, y_height_train, y_height_test = train_test_split(x_height, y_height, test_size = 0.3)
#Here, we impute values based on regression. 
#This is similar to mean imputation but this allows for different means based on demographics

#Imputing heights:
height_test_model = lm.fit(x_height_train, y_height_train)
height_test_predicted = height_test_model.predict(x_height_test)
rms = sqrt(mean_squared_error(y_height_test, height_test_predicted))
print('RMSE for Height: ', rms)
```

```python
# Evaluating Model Fit for imputation: Weights

x_weight = missing_naive.drop('weight', axis = 1)
y_weight = pd.DataFrame(missing_naive['weight'])
x_weight_train, x_weight_test, y_weight_train, y_weight_test = train_test_split(x_weight, y_weight, test_size = 0.3)
#Again, we impute values based on regression.

#Imputing weights:
weight_test_model = lm.fit(x_weight_train, y_weight_train)
weight_test_predicted = weight_test_model.predict(x_weight_test)
rms = sqrt(np.mean((y_weight_test - weight_test_predicted)**2))
print('RMSE: ', rms)
```

```python
#Next, we drop na values for people who have at least 2 missing values b/c they are hard to impute
indexes_membership = gym[gym['membership_type'].isnull()].index
for i in indexes_membership:
    if gym['age'][i] < 13:
        gym.at[i, ['membership_type']] = 'kids'
    elif gym['age'][i] < 18:
        gym.at[i, ['membership_type']] = 'youth'
    else:
        gym.at[i, ['membership_type']] = 'adults'
gym2 = gym.dropna(thresh = 6)

height_model = lm.fit(missing_naive.drop('height', axis = 1), missing_naive['height'])
print(height_model.coef_)
height_predict = height_model.predict(gym2[gym2['height'].isnull()][['age', 'sex', 'weight']])
indexes_height = gym2[gym2['height'].isnull()].index
for i in range(len(height_predict)):
    gym2.at[indexes_height[i], ['height']] = np.float64(height_predict[i])

#Imputing weights
weight_model = lm.fit(missing_naive.drop('weight', axis = 1), missing_naive['weight'])
print(weight_model.coef_)
weight_predict = weight_model.predict(gym2[gym2['weight'].isnull()][['age', 'sex', 'height']])
indexes_weight = gym2[gym2['weight'].isnull()].index
for i in range(len(weight_predict)):
    gym2.at[indexes_weight[i], ['weight']] = np.float64(weight_predict[i])
```

Using the naive method dataset, we are able to see many general patterns for predicting height and weight. These are height increses with age, men are generally heavier than women, and taller people tend to be slightly heavier. Thinking realistically, this makes sense as it is the case that older people are taller than younger people, men are taller than women, and people who are taller tend to weigh more, on average. While this correlation method I am using is helping to retain more data, the issue is that if both height and weight are missing from the dataset, neither are imputable via the regression method I decided on, so we cannot retain these observations od data. Thus, our final dataset, displayed below, is still not as large as the original dataset. However, I was still able to imputer missing values in a more sensible and robust fashion and retain more than five times as much data as the naive method. On top of this, we now have a much larger dataset with no missing values.
